# Contraste de hipótesis

## Teoría

La definición de las hipotesis es el cimiento lógico sobre el cual se construye todo el análisis estadístico. 

1. La Dualidad de la Definición: H0  frente a H1
​
Definir la hipótesis no consiste en realizar una única afirmación, sino en estructurar un conflicto entre dos proposiciones excluyentes:
La Hipótesis Nula (H0): Se define paradójicamente como la suposición inicial que el investigador pretende rechazar
. Es la base de la comparación y suele asociarse con signos de igualdad o estabilidad ("es igual a", "es menor o igual a")
. La Hipótesis Alternativa (H1 ): Es la verdadera "hipótesis de investigación"
. Representa la suposición previa del investigador y lo que este pretende probar como cierto
. Este contraste es fundamental porque la estadística no busca "confirmar" una idea directamente, sino verificar si la evidencia es suficiente para abandonar la suposición inicial (H0) en favor de la alternativa (H1)

2. De la Pregunta Narrativa a la Estructura Matemática
Las fuentes sugieren que la definición actúa como un puente entre una duda del mundo real y un modelo matemático.
La Pregunta: El proceso siempre parte de una pregunta que busca respuesta (por ejemplo: "¿son los chicos más altos que las chicas?")
.
La Formalización: La definición exige traducir esa pregunta a parámetros poblacionales específicos, como la media o la proporción
. Esta elección determina qué tipo de prueba (Z-test, T-test, ANOVA) será necesaria más adelante
.
3. El Margen de Error en la Definición
Un punto de debate interesante en las fuentes es que la definición de la hipótesis está intrínsecamente ligada al nivel de significación (α)
.
Al definir la hipótesis, el investigador también debe definir cuánto error está dispuesto a aceptar al rechazar una hipótesis nula que en realidad es verdadera
.
Las fuentes aclaran que no existe un nivel de significación del 0%, por lo que toda definición de "verdad" estadística conlleva un riesgo inherente de error o sesgo
.
4. La Importancia de la Población y la Muestra
Finalmente, la definición de una hipótesis es una afirmación respecto a una característica de una población
. Sin embargo, en la práctica, el contraste se realiza sobre una muestra significativa
. El debate aquí reside en la representatividad: la hipótesis se define para la población total (por ejemplo, 47 millones de españoles), pero se verifica mediante un subconjunto o muestreo, ya que analizar a toda la población suele ser inabordable
.
En conclusión, según las fuentes, definir en este contexto significa transformar una suposición subjetiva en una estructura matemática binaria (H 
0
​
  vs H 
1
​
 ), delimitando el parámetro a medir y el margen de error aceptable antes de observar la realidad de los datos

## Ejercicio 1

Usted es un nutricionista que investiga dos tipos diferentes de dietas para ver si existe una diferencia significativa en la pérdida de peso después de un mes. Eliges dos grupos aleatorios de personas; un grupo sigue la primera dieta y el otro sigue la segunda. Al final del mes, se registra la pérdida de peso (en kg) de cada persona:

| Diet 1 | Diet 2 |
|:-------|:-------|
| 2.0 | 3.0 |
| 2.5 | 3.2 |
| 3.0 | 3.1 |
| 2.8 | 2.9 |
| 2.3 | 2.8 |
| 2.7 | 3.0 |
| 2.5 | 3.2 |

Con estos datos se busca responder a la siguiente pregunta: ¿Existe una diferencia significativa en la pérdida de peso promedio entre las personas que siguieron la primera dieta y las que siguieron la segunda dieta?

Para sacar conclusiones, siga los siguientes puntos:

-Enunciar la hipótesis: hipótesis nula y alternativa.
-Realizar la prueba para comprobar la hipótesis. Puede utilizar la prueba t de Student.
-Analizar las conclusiones.

> **Hipótesis Nula (H0):** No existe una diferencia significativa en la pérdida de peso promedio entre la Dieta 1 y la Dieta 2. Esta es la suposición inicial que pretendemos rechaza.  
> **Hipótesis Alternativa (H1):** Existe una diferencia significativa en la pérdida de peso promedio entre ambas dietas. Esta es la hipótesis de investigación que queremos probar

```python
import scipy.stats as stats

diet_1 = [2.0, 2.5, 3.0, 2.8, 2.3, 2.7, 2.5]
diet_2 = [3.0, 3.2, 3.1, 2.9, 2.8, 3.0, 3.2]

# Student's t-test
t_value, p_value = stats.ttest_ind(diet_1, diet_2)

print(f"t-value: {t_value}")
print(f"p-value: {p_value}")
```

> **Conclusión:**  
> Para decidir si aceptamos o rechazamos la hipótesis, comparamos el p-valor con el nivel de significación (α), que normalmente se establece en 0.05 (5%). Como p-valor (0.00408) es mucho menor que 0.05, concluimos que los resultados son altamente significativos. Esto significa que es muy poco probable obtener estos datos si la Dieta 1 y la Dieta 2 fueran iguales.  
> **Decisión:** Rechazamos la hipótesis nula (H0) y aceptamos la hipótesis alternativa (H1). Existe evidencia estadística suficiente para afirmar que hay una diferencia significativa en la pérdida de peso entre las dos dietas.

## ANOVA

**ANOVA** (*Analysis of Variance - Análisis de variación*) Es una técnica estadística utilizada para comparar las medidas de dos o más grupos. La idea detrás de ANOVA es descomponer la variabilidad total de los datos en dos componentes: variabilidad entre grupos y variabilidad dentro de los grupos:

- **Variabilidad entre grupos**: Esta variabilidad se refiere a las diferencias entre las medias del grupo. Si esta variabilidad es considerablemente mayor que la variabilidad dentro del grupo, podría ser una indicación de que al menos una de las medias del grupo es diferente.
- **Variabilidad dentro del grupo**: Esta variabilidad se refiere a la dispersión de los datos dentro de cada grupo. Si todos los grupos tienen una variabilidad similar, entonces cualquier diferencia notable en las medias de los grupos podría considerarse significativa.

Las hipótesis en ANOVA suelen incluir:

- **Hipótesis nula** ($H₀$): Las medias de todos los grupos son iguales.
- **Hipótesis alternativa** ($H₁$): Al menos una de las medias del grupo es diferente.

Si el resultado de la prueba ANOVA es significativo (por ejemplo, un valor p inferior a un umbral como 0,05), esto sugiere que al menos la media de un grupo es diferente.

## Ejercicio 2

Un agricultor decide probar tres tipos diferentes de fertilizantes para determinar si alguno es superior en términos de producción de maíz. El agricultor planta maíz en 15 parcelas idénticas y utiliza los tres fertilizantes (5 parcelas para cada tipo). Al final de la temporada mide el rendimiento de maíz (en kg) de cada parcela, con el siguiente resultado:

| Fertilizante 1 | Fertilizante 2 | Fertilizante 3 |
|:-------------|:-------------|:-------------|
| 20 | 22 | 24 |
| 21 | 21 | 23 |
| 20 | 23 | 22 |
| 19 | 22 | 23 |
| 20 | 21 | 24 |

Con estos datos busca responder la siguiente pregunta: ¿Existe una diferencia significativa en el rendimiento promedio del maíz entre los tres tipos de fertilizantes?

Para ayudarte, sigue los siguientes puntos:

-Enunciar la hipótesis: hipótesis nula y alternativa.
-Realizar la prueba ANOVA.
-Analizar las conclusiones.
-Si un fertilizante es mejor que otro, ¿cómo podemos saberlo?

> **Hipótesis Nula (H0):** Las medias de rendimiento de maíz para los tres tipos de fertilizantes son iguales.  
> **Hipótesis Alternativa (H1):** Al menos una de las medias de rendimiento de los fertilizantes es diferente a las demás.

```python
import scipy.stats as stats

fertilizer_1 = [20,21,20,19,20]
fertilizer_2 = [22,21,23,22,21]
fertilizer_3 = [24,23,22,23,24]

f_value,p_value = stats.f_oneway(fertilizer_1,fertilizer_2,fertilizer_3)

print(f"f-value:{f_value}")
print(f"f-palue:{p_value}")
```

> **Observaciones:** El valor F indica que la variabilidad entre los grupos es considerablemente mayor que la variabilidad dentro de los grupos, lo que sugiere que el tipo de fertilizante sí influye en el resultado.


> **Conclusión:**  
> Nuevamente, observamos el p-valor. En este caso, 0.00014 es extremadamente bajo (menor a 0.05), lo que nos indica una significancia estadística muy alta.  
> **Decisión:** Rechazamos la hipótesis nula (H0). Podemos asegurar con un alto grado de confianza que no todos los fertilizantes producen el mismo rendimiento de maíz; al menos uno de ellos es significativamente diferente
